**PRE-PROCESSING**

```python
from tensorflow.keras.layers import TextVectorization 
```

**Tokenization**

```python
X = df['comment_text']
y = df[df.columns[2:]].values
print(X)  # X contains all the comments
```

```python
print(y)   # y contents all the labels
```

```python
df[df.columns[2:].values]  #It gives The File format that we can pss to the tancerflow deep learning models
```

```python
MAX_Words = 200000 # number of words
```

**Initializing text vectorization**

```python
vectorizer = TextVectorization(max_tokens=MAX_Words,
                               output_sequence_length=2000,
                               output_mode='int')
```

```python
type(X)
```

```python
type(X.values)
```

```python
vectorizer.adapt(X.values)
```

```python
vectorizer.get_vocabulary()
```

```python
vectorizer('hello')
```

```python
vectorized_text = vectorizer(X.values)
```

```python
vectorized_text
```

**Creating Tansorflow Data Pipeline**

```python
#map, chache, shuffle, batch, prefetch  from_tensor_slices, list_file
dataset = tf.data.Dataset.from_tensor_slices((vectorized_text, y))
dataset = dataset.cache()
dataset = dataset.shuffle(160000)
dataset = dataset.batch(16)
dataset = dataset.prefetch(8)
```

```python
batch_X, batch_y= dataset.as_numpy_iterator().next()
```

```python
batch_X  # batch_x is the vectorized text examples
```

```python
batch_X.shape
```

```python
batch_y #batch y is the vactorized labels examples
```

```python
batch_y.shape
```

**Traning Validation And Test partition**

```python
train = dataset.take(int(len(dataset)*.7))  #70% of dataset is assign to train
val = dataset.skip(int(len(dataset)*.7)).take(int(len(dataset)*.2)) #20% of remaning dataset is assign to validation after skipping the first&05 of it 
test = dataset.skip(int(len(dataset)*.9)).take(int(len(dataset)*.1)) #!0% is assign to the testing part
```

```python
len(train)#In batches
```

```python
len(val) #In batches
```

```python
len(test) #In batchhes
```

```python
train_generator= train.as_numpy_iterator()
```

```python
train_generator.next()
```

**CREATE LSTM AND CNN MODEL**

```python
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.layers import TextVectorization, Conv1D, MaxPooling1D, Flatten
from tensorflow.keras.layers import LSTM, Dropout, Bidirectional, Dense, Embedding, Conv1D, MaxPooling1D, Concatenate, Flatten
```

```python
# Create Sequential Model with both LSTM and CNN
model = tf.keras.Sequential()
model.add(Embedding(MAX_Words+1, 32))
```

```python
# Create CNN Branch as a separate model
cnn_branch = tf.keras.Sequential()
cnn_branch.add(Conv1D(64, 5, activation='relu'))
cnn_branch.add(MaxPooling1D(pool_size=4))
cnn_branch.add(Conv1D(64, 5, activation='relu'))
cnn_branch.add(MaxPooling1D(pool_size=4))
cnn_branch.add(Flatten())
```

```python
# Create Sequential Model with both LSTM and CNN
model = tf.keras.Sequential()
model.add(Embedding(MAX_Words+1, 32))
```

```python
# LSTM Branch
model.add(Bidirectional(LSTM(32, activation='tanh')))

```

```python

model.add(Dense(128, activation='relu'))
model.add(Dense(256, activation='relu'))
model.add(Dense(128, activation='relu'))

# Final layer
model.add(Dense(6, activation='sigmoid'))
```

```python
model.compile(loss='BinaryCrossentropy', optimizer='Adam')
```

```python
model.summary()
```

```python
# One-hot encode the target labels
encoder = OneHotEncoder(sparse=False)
y_encoded = encoder.fit_transform(y)
```

```python
history = model.fit(train, epochs = 1 ,validation_data = val)
```

**Make Prediction**

```python
input_text = vectorizer(['You are Moron.'])
```

```python
res = model.predict(input_text)
```

```python
toxicity_classes = ['toxic', 'severe_toxic', 'obscene', 'threat', 'insult', 'identity_hate']
for i, class_name in enumerate(toxicity_classes):
    print(f"{class_name}: {res[0][i]}")
```

**Evaluate Model**

```python
from tensorflow.keras.metrics import Precision, Recall, CategoricalAccuracy
```

```python
pre = Precision()
re = Recall()
acc = CategoricalAccuracy()
```

```python
for batch in test.as_numpy_iterator(): 
    # Unpack the batch 
    X_true, y_true = batch
    # Make a prediction 
    yhat = model.predict(X_true)
    
    # Flatten the predictions
    y_true = y_true.flatten()
    yhat = yhat.flatten()
    
    pre.update_state(y_true, yhat)
    re.update_state(y_true, yhat)
    acc.update_state(y_true, yhat)
```

```python
print(f'Precision: {pre.result().numpy()}, Recall:{re.result().numpy()}, Accuracy:{acc.result().numpy()}')
```
